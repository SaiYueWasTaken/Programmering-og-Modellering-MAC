```python
#Oppgave 1
from pylab import *

I = 3 #Antall smittede
a = 0.2 #Smitterate
t = 48 #Tid
populasjon = 157759
I_l = []
t_l = []

for i in range(t):
    I_l.append(I)
    t_l.append(i)
    I = I + a*I*(1-I/populasjon)
plot(t_l, I_l)
show()    
```

```python
#Oppgave 2
from pylab import *

populasjon = 157759
I = 1
a = 4/populasjon
t = 48
S = 0
U_l = []
t_l = []
S_l = []
for i in range(t):
    I = I + a*I*(1-I/populasjon)*S
    S = populasjon - I
    S_l.append(I)
    t_l.append(i)
    U_l.append(S)
plot(t_l, U_l)
plot(t_l, S_l)
title('Smittede og Antall ikke smittede')
xlabel("Antall uker")
ylabel("Populasjon")
show()  

```

```python
#Oppgave 3
from pylab import *

# Startbetingelser
N =  157759 # Antall individer
a = 4/N # Smitterate
b = 3.5 # Bedringsrate
t = 48 # Tid

I = 3 #Antall smittede
S = N - I #Antall ikke smittede
R = 0 #Antall friske

syke = [I]
mulige = [S]
friske= [R]
t_l = [0]

for i in range(1,t):
    S = S - a*S*I
    I = I + a*S*I - b*I
    R = R + b*I
    syke.append(I)
    mulige.append(S)
    friske.append(R)
    t_l.append(i)

plot(t_l, syke)
plot(t_l, mulige)
plot(t_l, friske)
show()    
```
